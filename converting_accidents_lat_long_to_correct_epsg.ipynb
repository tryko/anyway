# Converting the latitude and longitude to correct ESPG format

Currently these are just numbers that can be used on a folium map.<br>
The problem rises when we want to use LAMAS (למ"ס) Bicycle maps and street maps.<br>
These map use something called EPSG that project dots from a sphere to a 2D plane.<br>
The advantage of these maps is they allow to get distances in correct format (meters)<br>
Which allows us to calculate accidents per KM and more analysis.<br>

We need to convert accidents latitude and longitude to a shapely point and then give it the correct Projection (EPSG=32626)

```python
import matplotlib.pyplot as plt
from shapely.geometry import Point
import geopandas as gpd
import pandas as pd
import numpy as np
from IPython.display import Image, display
```

Loading all accidents data

```python
involved_markers_heb = pd.read_csv('./csv_tables/original_csv_tables/involved_markers_hebrew.csv')
```

Checking we get latitude with no na values

```python
filtered_df = involved_markers_heb.dropna(subset=['longitude', 'latitude'])

longitude = filtered_df.longitude 
latitude = filtered_df.latitude
```

```python
filtered_df.shape
```

```python
# Convert DataFrame to GeoDataFrame
involved_markers_heb_gdf = gpd.GeoDataFrame(
    filtered_df,
    geometry=[Point(xy) for xy in zip(longitude, latitude)]
)

involved_markers_heb_gdf
```

```python
involved_markers_heb_gdf = involved_markers_heb_gdf.set_crs(epsg=32636)
involved_markers_heb_gdf.crs
```

```python
involved_markers_heb_gdf.to_parquet('./csv_tables/original_csv_tables/involved_markers_heb_gdf.parquet')
```
