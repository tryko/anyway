### Combining all the population files

```python
pop_2014 = pd.read_csv('./csv_tables/israel_population_2014.csv')
pop_2015 = pd.read_csv('./csv_tables/israel_population_2015.csv')
pop_2016 = pd.read_csv('./csv_tables/israel_population_2016.csv')
pop_2017 = pd.read_csv('./csv_tables/israel_population_2017.csv')
pop_2018 = pd.read_csv('./csv_tables/israel_population_2018.csv')
pop_2019 = pd.read_csv('./csv_tables/israel_population_2019.csv')
pop_2020 = pd.read_csv('./csv_tables/israel_population_2020.csv')
pop_2021 = pd.read_csv('./csv_tables/israel_population_2021.csv')
pop_2022 = pd.read_csv('./csv_tables/israel_population_2022.csv')

pop_ls = [pop_2014, pop_2015,pop_2016,pop_2017,pop_2018,pop_2019,pop_2020,pop_2021, pop_2022]
```

```python
start_year = 2014
for pop in pop_ls:
    pop['pop_year'] = start_year
    start_year += 1

```

```python
all_pop = pd.concat(pop_ls)
all_pop
```
