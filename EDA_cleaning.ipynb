### Create date column

```python
# Convert columns to string type
involved_markers_heb_ta['accident_year'] = involved_markers_heb_ta['accident_year'].astype(str)
involved_markers_heb_ta['accident_month'] = involved_markers_heb_ta['accident_month'].astype(str).str.zfill(2)  # zero pad days, months, hours, and minutes
involved_markers_heb_ta['accident_day'] = involved_markers_heb_ta['accident_day'].astype(str).str.zfill(2)
involved_markers_heb_ta['accident_hour'] = involved_markers_heb_ta['accident_hour'].astype(str).str.zfill(2)
involved_markers_heb_ta['accident_minute'] = involved_markers_heb_ta['accident_minute'].astype(str).str.zfill(2)

# Create the new date column
involved_markers_heb_ta['accident_date'] = pd.to_datetime(
    involved_markers_heb_ta['accident_year'] + '-' + 
    involved_markers_heb_ta['accident_month'] + '-' + 
    involved_markers_heb_ta['accident_day'] + ' ' + 
    involved_markers_heb_ta['accident_hour'] + ':' + 
    involved_markers_heb_ta['accident_minute']
)
```
